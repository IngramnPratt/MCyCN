```python
import pandas as pd

# Generar el Data Frame
data={'Medidas de Corte':[495.44, 499.81, 498.62, 498.19, 490.72, 491.16, 498.41, 508.16, 499.06, 498.41, 498.29, 494.19, 497.57, 507.13, 495.63, 497.76, 491.07, 492.37, 505.06, 503.67, 502.91, 500.07, 495.59, 507.58, 498.66, 499.92, 506.81, 498.92, 498.40, 503.47, 495.09, 499.98, 505.28, 493.07, 498.59, 496.10, 498.42, 494.30, 505.73, 506.77, 493.88, 499.83, 501.99, 500.13, 502.03, 500.63, 498.41, 504.13, 500.33, 500.29, 501.53, 494.03, 488.03, 503.97, 505.37, 502.04, 495.14, 497.74, 511.73, 496.63,]}

df=pd.DataFrame(data)
# Muestra el Data Frame


print(df)
print('El total de cortes es:', df.size)
```

```python
#a)	Determina la media y la mediana e indica si la tendencia central del proceso es adecuada

#1.Determinar la media:
media=df.mean()
print('La media de los cortes es:',media[0] , "mm")

#2.Determinar la mediana
median=df.median()
print('La mediana de los cortes es:', median[0], 'mm')

#3.Determinar la diferencia entre la media y mediana
diferencia = media[0] - median[0]
print('La diferencia entre la media y la mediana de los cortes es:', diferencia, 'mm')


#4.Determinar los cortes dentro de los intervalos 495 y 505
filtered_data = df[(df['Medidas de Corte'] >= 495) & (df['Medidas de Corte'] <= 505)]
print('Hay', filtered_data.size, 'cortes entre los los intervalos 495 y 505')

```

```python
#b) Calcula la desviación estándar del proceso

std_dev=df.std()
print('La desviacion estandar es:',std_dev[0])
```

```python
#Determinar el rango mínimo y el rango máximo para elaborar el histograma

Min=df.min()
print('El rango mínimo es:',Min[0])

Max=df.max()
print('El rango máximo es:',Max[0])
```

```python
#c)	Obtén el histograma de las medidas e interpreta el gráfico resultante

import matplotlib.pyplot as plt

# Configurar tamaño del histograma
plt.figure(figsize=(8, 6))

# Generar histograma de 5 rangos, contenedores o barras. S
df["Medidas de Corte"].hist(bins=5, range=(488.03, 511.73), edgecolor='black')

# Agregar etiquetas y título
plt.xlabel("Medidas de Corte")
plt.ylabel("Frecuencia")
plt.title("Histograma de medidas de corte")

# Mostrar histograma
plt.show()
```

```python
# Generar histograma de 10 rangos, contenedores o barras.
df["Medidas de Corte"].hist(bins=10, range=(488.03, 511.73), edgecolor='black')

# Agregar etiquetas y título
plt.xlabel("Medidas de Corte")
plt.ylabel("Frecuencia")
plt.title("Histograma de medidas de corte")

# Mostrar histograma
plt.show()
```
